**Phần 1: Thống kê dữ liệu**

1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần

```python
print(df.sort_values(by = "DH1"))
```

2. Hãy sắp xếp dữ liệu điểm DH1 tăng dần theo giới tính

```python
print(df.sort_values(by = ["GT", "DH1"]))
```

3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT

```python
def Q1(g):
  return np.percentile(g, 25)
def Q2(g):
  return np.percentile(g, 50)
def Q3(g):
  return np.percentile(g, 75)
def std(g):
  return np.std(g)

pivot_table_KT = pd.pivot_table(data = df, index = 'KT', values = 'DH1', aggfunc = ['sum', 'mean', 'median', 'max', 'min', std, Q1, Q2, Q3])
print(pivot_table_KT)
```

4. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT và KV

```python
pivot_table_KT_KV = pd.pivot_table(data = df, index = ['KT', 'KV'], values = 'DH1', aggfunc = ['sum', 'mean', 'median', 'max', 'min', std, Q1, Q2, Q3])
print(pivot_table_KT_KV)
```

5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1,
Q2 và Q3 của DH1 theo KT, KV và DT

```python
pivot_table_KT_KV_DT = pd.pivot_table(data = df, index=['KT','KV','DT'], values = 'DH1', aggfunc = ['sum', 'mean', 'median', 'max', 'min', std, Q1, Q2, Q3])
print(pivot_table_KT_KV_DT)
```
